```python
from sympy import Symbol, lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = Symbol('x')
y = 5*x + 1

x_values = np.linspace(0, 10, 100)
y_values = lambdify(x, y, modules=['numpy'])(x_values)

plt.plot(x_values, y_values)
plt.ylabel("y = 5*x + 1")
plt.xlabel("x")
plt.show()
```

```python
from sympy import integrate, symbols
x = symbols('x')
y = 3*x**2 + 4*x - 5

area_under_curve = integrate(y, (x, 1, 3))
expression = integrate(y, x)
print(expression)
print(area_under_curve)
```
